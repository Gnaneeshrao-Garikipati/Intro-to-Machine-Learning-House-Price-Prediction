#### Apart from NA value columns converting other categorical columns into numerical, Using dummy variables method.

```python
# This get_dummies function when performed on a feature or column it treats each unique value as a column and creats a column
# with that name and gives one value if that value was present in that row if not zero is assigned.

dummies1 = pd.get_dummies(train.MSZoning, drop_first=True)
dummies1
```

```python
# Merging whatever the dummies we have created into the main train dataset.

merge1 = pd.concat([train, dummies1], axis='columns')
merge1
```

```python
# Droping the MSZoning column because as we have created dummies from that column and also merged them into the dataset we dont 
# have to include that column, if we include and do the other actions the predictions wont be up to the mark.

drop1 = merge1.drop(['MSZoning'], axis='columns')
drop1
```

```python
# As we see that Id column is basically the index so it does not posses any type of relationship with other columns so we 
# drop it.

drop_id = drop1.drop(['Id'], axis='columns')
drop_id
```

```python
# Trying the get_dummies function on other columns as well.

dummies2 = pd.get_dummies(drop_id.Street)
dummies2
```

```python
# Merging the dummy columns into the dataset.

merge2 = pd.concat([drop_id, dummies2], axis='columns')
merge2
```

```python
# Dropping the column from which the dummy columns have been formed.

train2 = merge2.drop(['Street'], axis='columns')
train2
```

```python
# Doing get_dummies function on another column.

dummies3 = pd.get_dummies(train2.LotShape)
dummies3
```

```python
# Merging the dummy columns into the dataset.

merge3 = pd.concat([train2, dummies3], axis='columns')
merge3
```

```python
# Dropping the column from which the dummy columns were formed.

train3 = merge3.drop(['LotShape'], axis='columns')
train3
```

#### test_size is the number that defines the size of the test set.

```python
# Importing Linear Regression to predict the saleprice in test data set.
# As the data which we have to predict is numerical so we use regression techniques, we are going to start with Linear Regression.

from sklearn.linear_model import LinearRegression
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# Predicting the saleprice values from the x_test which is test sub division of train dataset.

y_pred = regressor.predict(X_test)
y_pred
```

```python
# Importing and printing r2 score which tells us the accuracy of the model.

from sklearn.metrics import r2_score
r2_score(Y_test, y_pred)
```
