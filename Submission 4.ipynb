### Dividing the Dataset into Categorical and Numerical Datasets

```python
train_num=train.select_dtypes(exclude=['object'])
train_num
```

```python
train_cat=train.select_dtypes(exclude=['int64','float64'])
train_cat
```

### Filling the Missing Values

```python
sns.set_style("whitegrid")
missing = train_cat.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
plt.figure(figsize=(14,6))
ax=missing.plot.bar()
ax.bar_label(ax.containers[0])
```

```python
# Dropping unwanted columns
train_cat.drop(['PoolQC','Fence','Alley','MiscFeature','FireplaceQu'],axis=1,inplace=True)
```

```python
# Replacing categorical data with mode
train_cat['Electrical'].fillna(train_cat['Electrical'].mode(), inplace=True)
train_cat['MasVnrType'].fillna(train_cat['MasVnrType'].mode(), inplace=True)
train_cat['BsmtQual'].fillna(train_cat['BsmtQual'].mode(), inplace=True)
train_cat['BsmtCond'].fillna(train_cat['BsmtCond'].mode(), inplace=True)
train_cat['BsmtFinType1'].fillna(train_cat['BsmtFinType1'].mode(), inplace=True)
train_cat['BsmtFinType2'].fillna(train_cat['BsmtFinType2'].mode(), inplace=True)
train_cat['GarageType'].fillna(train_cat['GarageType'].mode(), inplace=True)
train_cat['GarageFinish'].fillna(train_cat['GarageFinish'].mode(), inplace=True)
train_cat['GarageQual'].fillna(train_cat['GarageQual'].mode(), inplace=True)
train_cat['GarageCond'].fillna(train_cat['GarageCond'].mode(), inplace=True)
train_cat['BsmtExposure'].fillna(train_cat['BsmtExposure'].mode(), inplace=True)
```

```python
# Replacing string categorical data with more frequent value
train_cat1 = train_cat.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
train_cat1.head()
```

```python
train_cat1.isnull().sum()
```

```python
sns.set_style("whitegrid")
missing = train_num.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
plt.figure(figsize=(14,6))
ax3=missing.plot.bar()
ax3.bar_label(ax3.containers[0])
```

```python
# Changing Datatype of GarageYrBlt to apply mode
train_num['GarageYrBlt'] = train_num['GarageYrBlt'].fillna(0).astype(int)
```

```python
# Filling columns with appropriate measure
train_num['LotFrontage'].fillna(train_num['LotFrontage'].mean(), inplace=True)
train_num['MasVnrArea'].fillna(train_num['MasVnrArea'].mean(), inplace=True)
train_num['GarageYrBlt'].fillna(train_num['GarageYrBlt'].mode(), inplace=True)
```

```python
train_num.isnull().sum()
```

### Replacing Categorical Data Into Numercial Data

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
cat=['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition']
train_cat1[cat]= train_cat1[cat].apply(label_encoder.fit_transform)
train_cat1
```

### Merging Categorical and Numerical Datasets

```python
frames=[train_cat1,train_num]
train1=pd.concat(frames, axis=1)
train1.info()
```
